### Visualization with Matplotlib

Matplotlib is easy to use and an amazing visualizing library in Python. It is built on NumPy arrays and designed to work with the broader SciPy stack and consists of several plots like line, bar, scatter, histogram, etc.

**Pyplot**

**Pyplot** is a Matplotlib module that provides a MATLAB-like interface. Pyplot provides functions that interact with the figure i.e. creates a figure, decorates the plot with labels, creates plotting area in a figure.

````python
# Python program to show pyplot module
import matplotlib.pyplot as plt
import numpy as np

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
````

**Python Matplotlib Bar Chart**

A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. The bar plots can be plotted horizontally or vertically. A bar chart describes the comparisons between the discrete categories. It can be created using the **bar()** method.

````python
fpath = 'data/'
````

````python
import matplotlib.pyplot as plt
import pandas as pd

# Reading the CSV file
df = pd.read_csv(fpath + 'input/iris.csv')

# This will plot a simple bar chart
plt.bar(df['species'], df['sepal_length'])

# Title to the plot
plt.title("Iris Dataset")

# Adding the legends
plt.legend(["count"])

# Adding axis labels
plt.xlabel("Species Type")
plt.ylabel("Sepal Length")

plt.show()
````

**Python Matplotlib Histogram**

A histogram is basically used to represent data in the form of some groups. It is a type of bar plot where the X-axis represents the bin ranges while the Y-axis gives information about frequency. To create a histogram the first step is to create a bin of the ranges, then distribute the whole range of the values into a series of intervals, and count the values which fall into each of the intervals. Bins are clearly identified as consecutive, non-overlapping intervals of variables. The **hist()** function is used to compute and create a histogram of x.

Following **example** plots a histogram of marks obtained by students in a class. Four bins, 0-25, 26-50, 51-75, and 76-100 are defined. The Histogram shows number of students falling in this range.

````python
from matplotlib import pyplot as plt
import numpy as np

marks = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])

plt.hist(marks, bins = [0,25,50,75,100])

# Title to the plot
plt.title("Student's Marks")

plt.xlabel('Marks')
plt.ylabel('Number of students')

plt.show()
````

## Installing new libraries in Python

As a popular open source development project, Python has an active supporting community of contributors and users that also make their software available for other Python developers to use under open source license terms.

This allows Python users to share and collaborate effectively, benefiting from the solutions others have already created to common (and sometimes even rare!) problems, as well as potentially contributing their own solutions to the common pool.

### Key terms

*   `pip` is the preferred installer program. 
*   A *virtual environment* is a semi-isolated Python environment that allows packages to be installed for use by a particular application, rather than being installed system wide.
*   `venv` is the standard tool for creating virtual environments, and has been part of Python since Python 3.3.
*   `virtualenv` is a third party alternative (and predecessor) to venv. It allows virtual environments to be used on versions of Python prior to 3.4.
*    The [Python Package Index](https://pypi.org/) is a public repository of open source licensed packages made available for use by other Python users.
*    `distutils` is the original build and distribution system first added to the Python standard library in 1998. While direct use of `distutils` is being phased out, it still laid the foundation for the current packaging and distribution infrastructure, and it not only remains part of the standard library, but its name lives on in other ways (such as the name of the mailing list used to coordinate Python packaging standards development).

### Basic usage

The standard packaging tools are all designed to be used from the command line.

The following command will install the latest version of a module and its dependencies from the Python Package Index:

`python -m pip install SomePackage`

It's also possible to specify an exact or minimum version directly on the command line. When using comparator operators such as >, < or some other special character which get interpreted by shell, the package name and the version should be enclosed within double quotes:

```
python -m pip install SomePackage==1.0.4    # specific version
python -m pip install "SomePackage>=1.0.4"  # minimum version
```

Normally, if a suitable module is already installed, attempting to install it again will have no effect. Upgrading existing modules must be requested explicitly:

`python -m pip install --upgrade SomePackage`

More information and resources regarding `pip` and its capabilities can be found in the [Python Packaging User Guide](https://packaging.python.org/en/latest/).

**Creating virtual environments:**

Creation of virtual environments is done through the `venv` module. Installing packages into an active virtual environment uses the commands shown above.

`python3 -m venv /path/to/new/virtual/environment`

## Data visualization with Plotly

**Why Plotly?**

There are numerous benefits to using Plotly. However, I want to draw your attention to a few important factors relevant to this course.

1. Most of data scientists and engineers write in Python. Their roles oftentimes require them to be able to aggregate small and big datasets fluently, to manipulate any kinds of data efficiently, and to work with libraries like `Pandas`. `Plotly` allows us to build charts for the web right from `Pandas` dataframe.
2. While Seaborn and Matplotlib graphics are static, `Plotly` charts allow you to **hover over values and zoom in/out your graphs**, like identifying outliers among a large number of data points or detecting anomalies in plots.
3. Plotly's Python graphing library makes **interactive, publication-quality graphs.**

Below reference contains examples of how to make line plots, scatter plots, area charts, bar charts, error bars, box plots, histograms, heatmaps, subplots, multiple-axes, polar charts, and bubble charts.

Reference: https://plotly.com/python/

````python
# Python program to show plotly module
import numpy as np
import plotly.express as px

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
print(len(x))
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
print(len(y))

fig = px.line(x=x,
              y=y,
              title='Health')

fig.update_layout(
    xaxis_title="Average Pulse",
    yaxis_title="Calorie Burnage"
)

fig.show()
````

**Run the below code and observe the output graph**

````python
import plotly.express as px

df = px.data.gapminder().query("continent=='Oceania'")
fig = px.line(df, x="year", y="lifeExp", color='country')
fig.show()
````

The above graph is too wide, and the y-axis label needs to be more user-friendly and have an intuitive name. Additionally, the x and y-axis labels and legend titles do not start with a capital letter. The graph does not have a title.

We can use the attributes below to customize the figure's `layout` a bit.

1. For `title`: `fig.update_layout(title="title of the graph")`
2. For `legend ` title: `fig.update_layout(legend_title="legend's title")`
3. For `width`: `fig.update_layout(width=<VALUE>)` # VALUE: number greater than or equal to 10
4. For `height`: `fig.update_layout(height=<VALUE>)` # VALUE: number greater than or equal to 10

For more refer https://plotly.com/python/reference/layout/

**Run the below code and observe the output graph**

````python
import plotly.express as px

df = px.data.gapminder().query("continent=='Oceania'")
fig = px.line(df, x="year", y="lifeExp", color='country')

# you need to update the layout after you plot the graph
fig.update_layout(title="Life expectancy in Canada",  # adding title to the graph
                  width=700,  # adjusting width of the graph
                  height=400,  # adjusting height of the graph
                  legend_title="Country",  # updating default legend's title
                  xaxis_title="Year",  # updating x-axis's title
                  yaxis_title="Life expectancy",  # updating y-axis's title
                  )

fig.show()
````

Now you can observe that the size and title got updated, and it also changed the labels of the x-axis, y-axis, and legend titles.

**Line charts with markers**

The `markers` argument can be set to `True` to show markers on lines.

````python
import plotly.express as px

df = px.data.gapminder().query("continent == 'Oceania'")
fig = px.line(df, x='year', y='lifeExp', color='country', markers=True)

fig.update_layout(title="Life expectancy in Canada",  # adding title to the graph
                  width=800,  # adjusting width of the graph
                  height=500,  # adjusting height of the graph
                  legend_title="Country",  # updating default legend's title
                  xaxis_title="Year",  # updating x-axis's title
                  yaxis_title="Life expectancy",  # updating y-axis's title
                  )

fig.show()
````

## Save the graph to the local drive as HTML file

Any figure can be saved as an HTML file using the `write_html` method. These HTML files can be opened in any web browser to access the fully interactive figure.

````python
import plotly.express as px

df = px.data.gapminder().query("continent == 'Oceania'")
fig = px.line(df, x='year', y='lifeExp', color='country', markers=True)

fig.update_layout(title="Life expectancy in Canada",  # adding title to the graph
                  width=800,  # adjusting width of the graph
                  height=500,  # adjusting height of the graph
                  legend_title="Country",  # updating default legend's title
                  xaxis_title="Year",  # updating x-axis's title
                  yaxis_title="Life expectancy",  # updating y-axis's title
                  )

fig.write_html(fpath + 'output/life_expectancy.html')
````
